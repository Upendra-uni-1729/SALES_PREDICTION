```python
df = df.loc[:,['TV','radio','newspaper','sales']]
```

```python
df.head()
```

```python
##Checking the linearty b/w the features and target

plt.scatter(df['TV'], df['sales'], color='blue', marker='o')  # TV vs sales

# Adding labels and title
plt.xlabel('TV')
plt.ylabel('sales')
plt.title('TV vs sales')
#saving the plot image
plt.savefig('./data_plots/TV vs Sales.jpg',dpi=300)
# Show the plot
plt.show()

plt.scatter(df['radio'], df['sales'], color='red', marker='x') # radio vs sales

# Adding labels and title
plt.xlabel('radio')
plt.ylabel('sales')
plt.title('radio vs sales')
#saving the plot image
plt.savefig('./data_plots/radio vs Sales.jpg',dpi=300)
# Show the plot
plt.show()

plt.scatter(df['newspaper'], df['sales'], color='green', marker='o') # newspaper vs sales

# Adding labels and title
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.title('newspaper vs sales')
#saving the plot image
plt.savefig('./data_plots/newspaper vs Sales.jpg',dpi=300)
# Show the plot
plt.show()
```

```python
# correlation b/w the features and target
correlation = df.corr()
     

# constructing a heatmap to nderstand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
plt.savefig('./data_plots/correlation_heatmap.jpg')

# on observing the correlation b/w features and target there is no feature with zero correlation
```
